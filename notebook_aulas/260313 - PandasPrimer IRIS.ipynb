## Step 3: Data cleaning (NaN handling)

### Some ML models can't handle null values.

```python
# Check for null values
print("Null values per column:\n", df.isnull().sum())
```

```python
# Strategy: fill NaNs with the mean of the column
df['petal_length'] = df['petal_length'].fillna(df['petal_length'].mean())
df['petal_width'] = df['petal_width'].fillna(df['petal_width'].mean())

print("\nNulls after processing:", df.isnull().sum().sum())
```

## Step 4: Transformation (Feature Engineering)

### ML algorithms work with math. Therefore they need numbers. Let's convert the "target" column.

```python
# Create the mapping dictionary
species_map = {
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}

# Apply the mapping
df['target_numeric'] = df['target'].map(species_map)

# Visualize the result
print("Sample of the transformation:")
print(df[['target', 'target_numeric']].sample(5))
```
